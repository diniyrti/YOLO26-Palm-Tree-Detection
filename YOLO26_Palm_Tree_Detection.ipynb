Split Dataset

```python
from sklearn.model_selection import train_test_split
import os, shutil
from pathlib import Path
```

```python
INPUT_IMAGES = "/content/drive/MyDrive/Linked_In_Content/yolo_detection/Palm Oil Detection.v1i.yolo26/train/images"
INPUT_LABELS = "/content/drive/MyDrive/Linked_In_Content/yolo_detection/Palm Oil Detection.v1i.yolo26/train/labels"
OUTPUT_DIR   = "/content/drive/MyDrive/Linked_In_Content/yolo_detection/Palm Oil Detection.v1i.yolo26/output_split"
TEST_SIZE    = 0.2
RANDOM_STATE = 42

image_files = [
    f for f in Path(INPUT_IMAGES).glob("*.jpg")
    if (Path(INPUT_LABELS) / (f.name.replace('.jpg', '.txt'))).exists()
]

print(f"Total pairs: {len(image_files)}")

train_files, val_files = train_test_split(image_files, test_size=TEST_SIZE, random_state=RANDOM_STATE)

for split, files in [("train", train_files), ("val", val_files)]:
    os.makedirs(f"{OUTPUT_DIR}/{split}/images", exist_ok=True)
    os.makedirs(f"{OUTPUT_DIR}/{split}/labels", exist_ok=True)
    for img in files:
        label = Path(INPUT_LABELS) / img.name.replace('.jpg', '.txt')
        shutil.copy(img,   f"{OUTPUT_DIR}/{split}/images/{img.name}")
        shutil.copy(label, f"{OUTPUT_DIR}/{split}/labels/{label.name}")

print(f"✅ Train: {len(train_files)} | Val: {len(val_files)}")
```

Running Model

```python
# Training model YOLO

from ultralytics import YOLO

# Load model YOLO
model = YOLO('/content/drive/MyDrive/Linked_In_Content/yolo_detection/yolo26m.pt')

# Start training
model.train(
    data='/content/drive/MyDrive/Linked_In_Content/yolo_detection/Palm Oil Detection.v1i.yolo26/output_split/data.yaml',
    epochs=100,
    imgsz=640,
    batch=8,
    project='/content/drive/MyDrive/Linked_In_Content/yolo_detection/running_hasil',
    name='yolo26m_PalmOil_detection',
    device=0,
    # select the augmentation you want to use
    flipud=0.5,      # objects do not have top-down orientation
    degrees=45,      # objects can face any direction
    hsv_h=0.0,       # maintain the color of vegetation

    # Stability training
    optimizer="AdamW",
    lr0=0.001,
    patience=20,
)
```
